```python
# Import pandas
import pandas as pd

# Define the path to your files
audio_metadata_path = '/content/drive/MyDrive/CSE425/Data/audio_metadata.csv'
lyrics_metadata_path = '/content/drive/MyDrive/CSE425/Data/lyrics_metadata.csv'

# Load the CSV files into pandas DataFrames
audio_metadata = pd.read_csv(audio_metadata_path)
lyrics_metadata = pd.read_csv(lyrics_metadata_path)

# Print the number of data points for audio metadata and lyrics metadata
print(f'Number of data points in audio_metadata: {audio_metadata.shape[0]}')
print(f'Number of data points in lyrics_metadata: {lyrics_metadata.shape[0]}')
```

```python
# Merge the audio and lyrics metadata on the 'AllMusic Id' column, keeping only rows that are in both datasets
genre = pd.merge(audio_metadata, lyrics_metadata, on='AllMusic Id')

# Print the number of data points and the first few rows
print(f'Number of data points: {genre.shape[0]}')
```

```python
# Print the number of columns and the column names in genre
print(f'Number of columns in genre: {genre.shape[1]}')
print(f'Column names in genre: {genre.columns.tolist()}')
```

```python
# Select the relevant columns with both 'x' and 'y' suffixes
genre = genre[['AllMusic Id', 'Title_x', 'Duration_x', 'Genres_x']]

# Rename the columns for clarity
genre.columns = ['Music_Id', 'Title', 'Duration', 'Genres']

# Show the first few rows
print(genre.head())
```

```python
# Save the merged DataFrame to a new CSV file named 'genre.csv'
genre.to_csv('/content/drive/MyDrive/CSE425/Data/genre.csv', index=False)
```
